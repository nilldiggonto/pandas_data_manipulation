```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set()
```

```python
plot_data = pd.read_csv('titanic.csv')
```

```python
plot_data.plot() #all numeric data
plt.show()
```

```python
plot_data.plot(subplots=True)
plt.show()
```

```python
plot_data.plot(subplots=True,figsize=(15,12))
plt.show()
```

```python
plot_data.plot(subplots=True,figsize=(15,12),sharex=False)
plt.show()
```

```python
plot_data['age'].plot(figsize=(10,10))
plt.show()
```

```python
#adding title and label
```

```python
plot_data['age'].plot(figsize=(12,8),fontsize=12,c='cyan',linestyle="-",xlim=(200,300),ylim=(20,50))
plt.title('Age Plot',fontsize=15)
plt.legend(loc='best',fontsize=14)
plt.xlabel('Passenger',fontsize=12)
plt.ylabel('Age',fontsize=12)
plt.grid()
plt.show()
```

```python
plot_data['age'].hist(figsize=(12,8),bins=5,xlabelsize=15,ylabelsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8)) #plt.hist method can't handle missing values but pandas method can
plt.hist(plot_data['age'].dropna(),bins=6,density=False,cumulative=False)
plt.show()
```
